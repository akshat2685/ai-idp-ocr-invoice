importing important libraries


```python
import pytesseract
from PIL import Image
import cv2
import json
import re
```

importing the ***INVOICE***



```python
from google.colab import files
uploaded = files.upload()
```

Here we are loading and processing the image

```python
image_path = list(uploaded.keys())[0]  # Get the actual uploaded filename, e.g., 'sample invoice image (1).webp'

img = cv2.imread(image_path)

# Check if image was loaded successfully
if img is None:
    raise FileNotFoundError(f"Error: Image not found at {image_path}. Please ensure the file exists and the path is correct.")

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Improve OCR quality
gray = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY)[1]

cv2.imwrite("processed.png", gray)
```

Extracting the text using OCR engine

```python
text = pytesseract.image_to_string(Image.open("processed.png"))
print(text)
```

Extracting important fields

```python
invoice_data = {}

# Invoice Number
invoice_no = re.search(r'Invoice\s*No[:\s]*([A-Z0-9-]+)', text, re.IGNORECASE)
invoice_data["invoice_number"] = invoice_no.group(1) if invoice_no else None

# Date
date = re.search(r'Date[:\s]*([0-9\/\-]+)', text)
invoice_data["date"] = date.group(1) if date else None

# Total Amount
total = re.search(r'Total[:\s₹]*([0-9,]+\.?[0-9]*)', text, re.IGNORECASE)
invoice_data["total_amount"] = total.group(1) if total else None

invoice_data
```

converting to JSON output

```python
json_output = json.dumps(invoice_data, indent=4)
print(json_output)
```

## Accuracy Improvements using NLP / LLMs

1. Post-OCR correction can be applied to fix spelling errors using NLP-based spell checkers.
2. Named Entity Recognition (NER) models can identify fields like invoice number, date, and total amount more reliably.
3. Large Language Models (LLMs) can be used to convert unstructured OCR text into structured JSON even when invoice formats vary.
4. Confidence scoring and validation rules can be added to detect incorrect extractions.
